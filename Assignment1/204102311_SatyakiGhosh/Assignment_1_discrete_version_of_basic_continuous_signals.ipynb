```
import matplotlib.pyplot as plt
import numpy as np
import math

zeroSamples=10
numSamples=10
rnge=10
n=np.arange(-rnge,rnge+1,1)


# UNIT STEP 
#t=np.linspace(-10,10,zeroSamples+numSamples)
x1=[]
for i in n:
  if(i<0):
    x1.append(0)
  else:
    x1.append(1)


#UNIT RAMP
x2=[0 for i in range(zeroSamples)]
for i in list(np.arange(0,numSamples+1,1)):
  x2.append(i)

#UNIT PARABOLIC
x3=[0 for i in range(zeroSamples)]
for i in list(np.arange(0,numSamples+1,1)):
  x3.append((i**2)/2)

#RECTANGULAR PULSE
#t2=np.linspace(-2,2,zeroSamples+numSamples)
x4=[0 for i in range(zeroSamples-1)]
for i in range(3):
  x4.append(1)
for i in range(numSamples-1):
  x4.append(0) 

#TRIANGULAR PULSE
#t2=np.linspace(-2,2,zeroSamples+numSamples)
x5=[0 for i in range(zeroSamples-3)]
x5.append(0.25)
x5.append(0.5)
x5.append(0.75)
x5.append(1)
x5.append(0.75)
x5.append(0.5)
x5.append(0.25)
for i in range(numSamples-3):
  x5.append(0) 

# SINC FUNTION
#t3=np.linspace(-6,6,zeroSamples+numSamples)
nForsinc=np.arange(-rnge,rnge+1,0.2)
x6=np.sinc(nForsinc)
#x6 = np.sin(math.pi * n) / (math.pi * n)


#PLOTTING THE BASIC FUNCTIONS
plt.rcParams.update({'font.size': 15})
fig,ax=plt.subplots(2,3,figsize=(20,10))
#plt.figure(figsize=(10,10))

ax[0,0].stem(n,x1,label='unit step function',use_line_collection=True)
ax[0,0].set_xlabel('time(samples)')
ax[0,0].legend()
ax[0,0].grid()
ax[0,1].stem(n,x2,label='unit ramp function',use_line_collection=True)
ax[0,1].set_xlabel('time(samples)')
ax[0,1].legend()
ax[0,1].grid()
ax[0,2].stem(n,x3,label='unit parabolic function',use_line_collection=True)
ax[0,2].set_xlabel('time(samples)')
ax[0,2].legend()
ax[0,2].grid()
ax[1,0].stem(n,x4,label='rectangular pulse function [-1,1]',use_line_collection=True)
ax[1,0].set_xlabel('time(samples)')
ax[1,0].legend()
ax[1,0].grid()
ax[1,1].stem(n,x5,label='triangular pulse function [-1,1]',use_line_collection=True)
ax[1,1].set_xlabel('time(samples)')
ax[1,1].legend()
ax[1,1].grid()
ax[1,2].stem(nForsinc,x6,label='sinc function',use_line_collection=True)
ax[1,2].set_xlabel('time(samples)')
ax[1,2].legend()
ax[1,2].grid()

fig.suptitle('Basic signals')

plt.show()

########################################################################
'''
zeroSamples=800
numSamples=1200

# UNIT STEP 
t=np.linspace(-10,10,zeroSamples+numSamples)
x1=[0 for i in range(zeroSamples)]
for i in range(numSamples):
  x1.append(1)


#UNIT RAMP
x2=[0 for i in range(zeroSamples)]
for i in list(np.linspace(0,12,numSamples)):
  x2.append(i)


#UNIT PARABOLIC
x3=[0 for i in range(zeroSamples)]
for i in list(np.linspace(0,12,numSamples)):
  x3.append((i**2)/2)
'''
x1=[5*i for i in x1]
for i in range(2):
  x1.pop()
  x1=[0]+x1
x2=[5*i for i in x2]
for i in range(2):
  x2.pop()
  x2=[0]+x2
x3=[5*i for i in x3]
for i in range(2):
  x3.pop()
  x3=[0]+x3

#PLOTTING THE BASIC FUNCTIONS
plt.rcParams.update({'font.size': 15})
fig,ax=plt.subplots(1,3,figsize=(30,8))
#plt.figure(figsize=(10,10))




ax[0].stem(n,x1,label='step function',use_line_collection=True)
ax[0].set_xlabel('samples')
ax[0].legend()
ax[0].grid()
ax[1].stem(n,x2,label='ramp function',use_line_collection=True)
ax[1].set_xlabel('samples')
ax[1].legend()
ax[1].grid()
ax[2].stem(n,x3,label='parabolic function',use_line_collection=True)
ax[2].set_xlabel('samples')
ax[2].legend()
ax[2].grid()

fig.suptitle('Amplitude-scaled (by 5) and time-shifted (by 2) Basic signals')
plt.show()

#########################################################################################
'''
zeroSamples=1000
numSamples=1000

# TIME SCALED BY 0.5

# UNIT STEP 
t=np.linspace(-10,10,zeroSamples+numSamples)
x1=[0 for i in range(zeroSamples)]
for i in range(numSamples):
  x1.append(1)


#UNIT RAMP
x2=[0 for i in range(zeroSamples)]
for i in list(np.linspace(0,5,numSamples)):
  x2.append(i)


#UNIT PARABOLIC
x3=[0 for i in range(zeroSamples)]
for i in list(np.linspace(0,5,numSamples)):
  x3.append((i**2)/2)
'''
# UPSAMPLING BY 2
# UNIT STEP 
#t=np.linspace(-10,10,zeroSamples+numSamples)
x1=[]
for i in n:
  if(i<0):
    x1.append(0)
  else:
    if(i%2==0):
      x1.append(1)
    else: 
      x1.append(0)


#UNIT RAMP
x2=[0 for i in range(zeroSamples)]
for i in list(np.arange(0,numSamples+1,1)):
  if(i%2==0):
    x2.append(i/2)
  else: 
    x2.append(0)

#UNIT PARABOLIC
x3=[0 for i in range(zeroSamples)]
for i in list(np.arange(0,numSamples+1,1)):
  if(i%2==0):
    x3.append(((i/2)**2)/2)
  else:
    x3.append(0)


#PLOTTING THE BASIC FUNCTIONS
plt.rcParams.update({'font.size': 15})
fig,ax=plt.subplots(1,3,figsize=(30,8))
#plt.figure(figsize=(10,10))

ax[0].stem(n,x1,label='step function',use_line_collection=True)
ax[0].set_xlabel('samples')
ax[0].legend()
ax[0].grid()
ax[1].stem(n,x2,label='ramp function',use_line_collection=True)
ax[1].set_xlabel('samples')
ax[1].legend()
ax[1].grid()
ax[2].stem(n,x3,label='parabolic function',use_line_collection=True)
ax[2].set_xlabel('samples')
ax[2].legend()
ax[2].grid()

fig.suptitle('Timescaled (by 0.5) Basic signals')

plt.show()


# TIME SCALED BY 2

# UNIT STEP 
#t=np.linspace(-10,10,zeroSamples+numSamples)
x1=[0 for i in range(zeroSamples)]
for i in range(11):
  x1.append(1)


#UNIT RAMP
x2=[0 for i in range(zeroSamples)]
for i in range(11):
  x2.append(2*i)


#UNIT PARABOLIC
x3=[0 for i in range(zeroSamples)]
for i in range(11):
  x3.append(((2*i)**2)/2)


#PLOTTING THE BASIC FUNCTIONS
plt.rcParams.update({'font.size': 15})
fig,ax=plt.subplots(1,3,figsize=(30,8))


ax[0].stem(n,x1,label='step function',use_line_collection=True)
ax[0].set_xlabel('samples')
ax[0].legend()
ax[0].grid()
ax[1].stem(n,x2,label='ramp function',use_line_collection=True)
ax[1].set_xlabel('samples')
ax[1].legend()
ax[1].grid()
ax[2].stem(n,x3,label='parabolic function',use_line_collection=True)
ax[2].set_xlabel('samples')
ax[2].legend()
ax[2].grid()

fig.suptitle('Timescaled (by 2) Basic signals')

plt.show()  

########################################################################

A= [-5,5]
B= [ -5, -2, -0.5, -0.25, 0, 1, 2]
rnge=10
t=np.arange(-rnge,rnge,1)
plt.rcParams.update({'font.size': 15})

for a in A: 
  for b in B: 
    x=a*np.exp(b*t)
    plt.figure(figsize=(10,4))
    plt.stem(t,x,use_line_collection=True)
    plt.xlabel('time (samples)')
    plt.ylabel('signal value')
    plt.title(f'DiscreteExponential signal : A={a}, B={b}')
    plt.grid()
    plt.show()

  ########################################################################

rnge=10
t=np.arange(-rnge,rnge,1)

# 5 sin (2 pi t)
x1=[0]*20
plt.figure(figsize=(10,6))
plt.stem(t,x1,use_line_collection=True)
plt.xlabel('time (samples)')
plt.ylabel('signal value')
plt.title(f'Sinusoidal signal : 5sin(2 $\pi$ n)')
plt.grid()
plt.show()

# 2 sin (2/3 pi t)
x2=2*np.sin((2/3)*np.pi*t)
plt.figure(figsize=(10,6))
plt.stem(t,x2,use_line_collection=True)
plt.xlabel('time (samples)')
plt.ylabel('signal value')
plt.title(f'Sinusoidal signal : 2sin(2/3 $\pi$ n)')
plt.grid()
plt.show()

# 4 cos (1/3 pi t)
x3=4*np.cos((1/3)*np.pi*t)
plt.figure(figsize=(10,6))
plt.stem(t,x3,use_line_collection=True)
plt.xlabel('time (samples)')
plt.ylabel('signal value')
plt.title(f'Cosinusoidal signal : 4cos($\pi$ /3 n)')
plt.grid()
plt.show()

# 3 cos (2 pi t + pi/3)
x4=3*np.cos((2*np.pi*t) + (np.pi/3))
plt.figure(figsize=(10,6))
plt.stem(t,x4,use_line_collection=True)
plt.xlabel('time (samples)')
plt.ylabel('signal value')
plt.title(f'Cosinusoidal signal : 3cos(2$\pi$n + $\pi$/3)')
plt.grid()
plt.show()

# x1 + x2 
x5=x1+x2
plt.figure(figsize=(10,6))
plt.stem(t,x5,use_line_collection=True)
plt.xlabel('time (samples)')
plt.ylabel('signal value')
plt.title(f'Sinusoidal signal : 5sin(2 $\pi$ n) +  2sin(2/3 $\pi$ n)')
plt.grid()
plt.show()

# x3 + x4
x6=x3+x4
plt.figure(figsize=(10,6))
plt.stem(t,x6,use_line_collection=True)
plt.xlabel('time (samples)')
plt.ylabel('signal value')
plt.title(f'Sinusoidal signal : 4cos($\pi$ /3 n) + 3cos(2$\pi$n + $\pi$/3)')
plt.grid()
plt.show()

# x1 + x3
x7=x1+x3
plt.figure(figsize=(10,6))
plt.stem(t,x7,use_line_collection=True)
plt.xlabel('time (samples)')
plt.ylabel('signal value')
plt.title(f'Sinusoidal signal : 5sin(2 $\pi$ n) + 4cos($\pi$ /3 n)')
plt.grid()
plt.show()

# x1(-t)
x8=[0]*20
plt.figure(figsize=(10,6))
plt.stem(t,x8,use_line_collection=True)
plt.xlabel('time (samples)')
plt.ylabel('signal value')
plt.title(f'Sinusoidal signal : 5sin(- 2$\pi$ n)')
plt.grid()
plt.show()

# x3(-t)
x9=4*np.cos(-(1/3)*np.pi*t)
plt.figure(figsize=(10,6))
plt.stem(t,x9,use_line_collection=True)
plt.xlabel('time (samples)')
plt.ylabel('signal value')
plt.title(f'Sinusoidal signal : 4cos(-$\pi$ /3 n)')
plt.grid()
plt.show()

################################################################################

pie=np.pi
A= [0.1, 0.5, 1, 2]
B= [ -0.25, -0.5, -1, 0.5, 1 ]
omega = [2*pie, pie/6 , 5*pie/3]

rnge=10
t=np.arange(-rnge,rnge,1)
plt.rcParams.update({'font.size': 15})

for a in A: 
  for b in B: 
    for w in omega:
      x=a*np.exp(b*t)*np.cos(w*t)
      plt.figure(figsize=(10,4))
      plt.stem(t,x,use_line_collection=True)
      plt.xlabel('time (samples)')
      plt.ylabel('signal value')
      plt.title(f'Discrete Exponentially varying sinusoidal signal : {a}*exp({b}t)*cos({w}t)')
      plt.grid()
      plt.show()

###################################################################################

pie=np.pi
params=[(1, 0.14*pie, 0) , (1, 0.14*pie, pie) ,(2, 0.34*pie, 0.5*pie) , (2, 0.75*pie, pie/3)]
rnge=10
t=np.arange(-rnge,rnge,1)
plt.rcParams.update({'font.size': 15})

for param in params:
  a=param[0]
  w=param[1]
  theta=param[2]
  real=a*np.cos(w*t + theta)
  imag=a*np.sin(w*t + theta)
  fig,ax=plt.subplots(1,2,figsize=(10,4))
  ax[0].stem(t,real,label='real part',use_line_collection=True)
  ax[0].set_xlabel('time (samples)')
  ax[0].grid()
  ax[0].legend()
  ax[1].stem(t,imag,label='imaginary part',use_line_collection=True)
  ax[1].set_xlabel('time (samples)')
  ax[1].grid()
  ax[1].legend()
  fig.suptitle(f'Discrete Complex Exponential signal : {a}*exp(j{w} t+{theta})')
  plt.show()


```
